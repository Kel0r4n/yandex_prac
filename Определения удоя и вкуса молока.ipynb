### Построим графики по соотношению различных категориальных признаков в основном стаде фермера

```python
def pie_plots(df, axis, about): #функция для построение pie графиков по категориальным признакам, с аргументами:
    lam = df[axis].value_counts()  #df - табл с данными, axis - столбец, по которому строится соотношение, 
    df[axis].value_counts().plot(  #about - название графика
        kind = 'pie', 
        y = lam, 
        autopct = '%1.0f%%',
        ylabel = "",
        title = 'Процентое соотношение по признаку {0}'.format(about),
        figsize=(5,5));
    plt.show()
```

## 4. Корреляционный анализ

### Взаимосвязь между признаками: коэффициенты корреляции между всеми признаками.

```python
main.phik_matrix(interval_cols = ['удой_кг', 'эке', 'сырой_протеин_г', 'спо', 'жирность', 'белок'])
```

```python
main1 = main.drop('id', axis = 1)
```

```python
display(sns.heatmap(main1.corr(method = 'spearman'), annot=True, fmt='.2f'));
```

Из тепловой карты видим, что наибольшее влияние среди количественных признаков оказывают ЭКЕ, СПО и сырой протеин

```python
for c in list(main[['порода', 'порода_папы_быка', 'вкус_молока', 'возраст', 'тип_пастбища']]):
    sns.scatterplot(data=main, x='эке',y='удой_кг', hue = c) #строим скаттерплоты для изучения влияния различных факторов 
    plt.title('Зависимость удоя от ЭКЕ и категориального признака {0}'.format(c)) #на удой с учетом категориальных признаков
    plt.ylabel('Удой, кг')
    plt.xlabel('Энергетическая кормовая единица')
    plt.show()
```

Видим, что рост ЭКЕ приводит к увеличению удоя с коровы, при этом:
- Породы практически никак не влияют на удой, обе породы при одинаковом откорме будут давать примерно одинаково молока
- Порода папы быка как раз влияет на удой, так, коровы, чьим отцом были быки породы Айдиал будут давать больше молока
- Преимущественно, коровы, в корме которых было больше ЭКО будут давать молоко вкуснее
- Возраст коровы определенно влияет на удой, так, коровы возрастом более 2 лет, будут давать намного больше молока, чем молодые
- Тип пастбища практически не влияет на удой от коровы, но все же можно сказать, что коровы пасущиеся на холмистых пастбищах будут давать больше молока

```python
for c in list(main[['порода', 'порода_папы_быка', 'вкус_молока', 'возраст', 'тип_пастбища']]):
    sns.scatterplot(data=main, x='спо',y='удой_кг', hue = c)
    plt.title('Зависимость удоя от СПО и категориального признака {0}'.format(c))
    plt.ylabel('Удой, кг')
    plt.xlabel('СПО')
    plt.show()
```

```python
main['спо'].unique()
```

Из графика видно, что при СПО ниже 0.9 удой колеблится от 5250 до 6000, а при спо > 0.92 от 6000 до 7500, т.е, если давать корове корм с спо > 0.9, то она будет давать больше молока, при этом:
- Порода практически не влияет на удой с коровы, возможно, РефлешнСоверинг дает совсем чуть больше молока
- Порода папы быка Айдиал будет давать больше молока, чем Соверин
- Коровы, в чьих кормах, содержание СПО выше 0.92 будут давать более вкусное молоко
- "Старшие" коровы будут давать больше молока
- Коровы пасиущиеся в холмистых местах будут давать чуть больше молока

```python
for c in list(main[['порода', 'порода_папы_быка', 'вкус_молока', 'возраст', 'тип_пастбища']]):
    sns.scatterplot(data=main, x='сырой_протеин_г',y='удой_кг', hue = c)
    plt.title('Зависимость удоя от сырой протеин и категориального признака {0}'.format(c))
    plt.ylabel('Удой, кг')
    plt.xlabel('Сырой протеин, г')
    plt.show()
```

Можно сказать, что при содержании протеина от 1700 до 2000 г, коровы будут давать удой одинаково независимо, они могут выдавать как 5250 кг, так и 6750, при содержании протеина от 2100 г ситуация чуток нормализуется и коровы начинают выдавать удой в среднем 6500кг. Влияние категориальных признаков выражено также почти как и во всех предыдущих случаях,те:

```python
X_train_ohe4 = encoder.fit_transform(X4_train[cat_col_names4])
X_test_ohe4 = encoder.transform(X4_test[cat_col_names4])

encoder_col_names4 = encoder.get_feature_names()
```

```python
X_train_ohe4 = pd.DataFrame(X_train_ohe4, columns=encoder_col_names4)
X_test_ohe4 = pd.DataFrame(X_test_ohe4, columns=encoder_col_names4)

X4_train_scaled = pd.DataFrame(X4_train_scaled, columns=num_col_names4)
X4_test_scaled = pd.DataFrame(X4_test_scaled, columns=num_col_names4)

model_4_train = pd.concat([X_train_ohe4, X4_train_scaled], axis=1)
model_4_test = pd.concat([X_test_ohe4, X4_test_scaled], axis=1)
```

### Анализ модели логистической регрессии

Ошибка первого рода намного критичнее для феремера, тк при большом количестве таких ошибок, фермер купит коров, которые должны давать вкусное молоко, а на деле будут давать не вкусное, когда главным показателем для него является как раз таки вкусное молоко.

- Precision модели с дефолтным порогом равен 0.5, т.е модель присваивает объектам класс 1 с точность 0.5, это является плохим показателем
- Recall равна 0.308, т.е модель в 31% случаев присвает 1 класс объектам этого класса.

Метрика precision намного важнее для фермера, тк купленная корова, которая будет давать не вкусное молоко, принесет больше убытков, чем убытки от упущенной выгоды, из-за некупленной коровы класса 1, из-за того что ей был присвоен класс 0
